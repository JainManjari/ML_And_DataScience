```python
def balanced(str,co,cc,n):
    if cc==n:
        print(str)
        return
    if co>cc:
        balanced(str+"}",co,cc+1,n)
    if co<n:
        balanced(str+"{",co+1,cc,n)
        
n=int(input())
balanced("",0,0,n)
```

```python
#quick sort

def quicksort(arr):
    if len(arr)<=1:
        return arr
    pivot=arr[0]
    left=[x for x in arr if x<pivot]
    middle=[x for x in arr if x==pivot]
    right=[x for x in arr if x>pivot]
    return quicksort(left)+middle+quicksort(right)
```

```python
print(quicksort([3,4,0,-1,10,9,-11]))
```
