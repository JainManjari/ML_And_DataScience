```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv("X.csv")
dfy=pd.read_csv("Y.csv")
```

```python
X=dfx.values
Y=dfy.values

X2=(X-X.mean())/X.std()
```

```python
plt.figure()
plt.scatter(X,Y,label="Raw Data")
plt.scatter(X2,Y,color="red",label="Normalised Data")
plt.legend()
plt.show()
```
