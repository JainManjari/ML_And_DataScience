## Using KKN

```python
def KKN(X,Y,test_point,K=5):
    res=[]
    
    for i in range(X.shape[0]):
        x=X[i]
        
        dist=(np.sum((x-test_point)**2)**0.5)
        res.append((dist,Y[i]))
    
    res=sorted(res)
    res=np.array(res[:K])
    labels=np.unique(res[:,1],return_counts=True)
    ind=np.argmax(labels[1]) 
    return int(labels[0][ind])
```

```python
xtest=xtest.values

#xtest = min_max_scaler.fit_transform(xtest)

print(xtest.shape)
print(X.shape)
print(Y.shape)
```

```python
res=[]
for i in range(xtest.shape[0]):
    x=xtest[i]
    a=KKN(X,Y,x,K=17) #17
    res.append(a)
    
res=pd.DataFrame(res)
```

```python
res.columns=["Outcome"]
res
```

```python
res.to_csv("diabetes_result.csv",index=False)
```

# Using Gradient Descent

```python
# X : Matrix of m*n
# x : vector : containing single example with n feautures

def hypo(X,theta):
    return np.dot(X,theta)


def predict(X,theta):
    yp=hypo(X,theta) 
    return yp


def score(Y,Yp):
    n=np.sum((Y-Yp)**2)
    d=np.sum((Y-Y.mean())**2)
    return (1-(n/d))*100
    
    

def error(X,y,theta): #cost
    yp=hypo(X,theta)
    e=np.sum((y-yp)**2)
    return e/X.shape[0]



def gradient(X,y,theta):
    M,N=X.shape
    grad=np.zeros((N,))
    
    yp=hypo(X,theta)
    grad=np.dot(X.T,yp-y)
    return grad/M


def gradient_descent(X,y,lr=0.1,iterations=300):
    M,N=X.shape
    theta=np.zeros((N,))
    error_list=[]
    
    for i in range(iterations):
        e=error(X,y,theta)
        error_list.append(e)
        
        grad=gradient(X,y,theta)
        
        theta=theta-lr*grad
            
    return theta,error_list
```

```python
dfx=pd.read_csv("Diabetes_XTrain.csv")
dfx
```

```python
dfy=pd.read_csv("Diabetes_YTrain.csv")
dfy
```

```python
X=dfx.values
Y=dfy.values.reshape((-1,))

Y.shape
```

```python
#Preprocessing

ones=np.ones((X.shape[0],1))
X=np.hstack((ones,X))
pd.DataFrame(X)
```
