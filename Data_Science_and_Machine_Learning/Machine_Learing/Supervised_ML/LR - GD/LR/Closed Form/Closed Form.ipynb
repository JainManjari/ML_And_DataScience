```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression

plt.style.use("seaborn")
```

```python
#Generate Dataset
X,Y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)

Y=Y.reshape(-1,1)
Y.shape,X.shape
```

```python
#Normalise

X=X-X.mean()/X.std()
```

```python
#Visualize

plt.figure()
plt.scatter(X,Y)
plt.show()
```

```python
ones=np.ones((X.shape[0],1))
X_new=np.hstack((ones,X))
X_new.shape
```

```python
X_new[:5,:]
```

```python
def predict(X,theta):
    return np.dot(X,theta)


def getThetaClosed(X,Y):
    Y=np.mat(Y)
    f=np.dot(X.T,X)
    s=np.dot(X.T,Y)
    theta=np.linalg.pinv(f)*s
    return theta
```

```python
theta=getThetaClosed(X_new,Y)
theta
```

```python
Y_prd=predict(X_new,theta)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X,Y_prd,c="red")
plt.show()
```
