```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:/Python/DataScieneML/Supervised ML/LR/Diabetes/training_diabetes.csv")
df
```

```python
data=df.values
data
```

```python
X=data[:,0:10]
Y=data[:,10]
```

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
alg=LR()
```

```python
alg.fit(X,Y)
```

```python
Y_train_predict=alg.predict(X)
```

```python
print(alg.score(X,Y_train_predict))
```

```python
test=pd.read_csv("D:/Python/DataScieneML/Supervised ML/LR/Diabetes//test_diabetes.csv")
test
```

```python
X_test=test.values
```

```python
Y_test_predict=alg.predict(X_test)

Y_test_predict=np.round(Y_test_predict,5) #round to five digits

test_df=pd.DataFrame(Y_test_predict)

test_df.to_csv("D:/Python/DataScieneML/Supervised ML/LR/Diabetes/test_predict_diabetes.csv",index=False,header=('Diabetes'))
```

```python
print(alg.score(X_test,Y_test_predict))
```
