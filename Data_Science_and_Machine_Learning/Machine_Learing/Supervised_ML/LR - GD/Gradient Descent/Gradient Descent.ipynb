```python
def step_gradient(points,learning_rate,m,c):
    m_slope=0
    c_slope=0
    M=len(points)
    for i in range(M):
        x=points[i,0]
        y=points[i,1]
        m_slope+=(-2/M)*(y-m*x-c)*(x)
        c_slope+=(-2/M)*(y-m*x-c)
    m1=m-learning_rate*(m_slope)
    c1=c-learning_rate*(c_slope)
    return m1,c1
```

```python
def cost(points,m,c):
    M=len(points)
    total_cost=0
    for i in range(M):
        x=points[i,0]
        y=points[i,1]
        ans=y-m*x-c
        ans**=2
        total_cost+=ans
    
    return total_cost/M
        
```

```python
def predict(points,m,c):
    M=len(points)
    y=[]
    for i in range(M):
        y.append(points[i,0]*m+c)
    return y
        
        
```

```python
def gd(points,learning_rate,num_iterations):
    m=0
    c=0
    for i in range(num_iterations):
        m,c=step_gradient(points,learning_rate,m,c)
        print(i,"Cost ",cost(points,m,c))
    return m,c
```

# Genric Gradient Descent

```python
def step_gradient(points,learning_rate,m):
    m=np.array(m)
    m_slope=np.zeros(m.shape[0],dtype=int)
    new_m=[]
    M=points.shape[0]
    N=points.shape[1]
    for i in range(M):
        y=points[i,N-1]
        sum=0
        for j in range(N-1):
            sum+=m[j]*points[i,j]
            
        for k in range(N-1):
            m_slope[k]+=(-2/M)*(y-sum)*points[i,k] 
            
    
    for i in range(m.shape[0]):
        new_m.append(m[i]-learning_rate*m_slope[i])
    
    return new_m
```

```python
def predict(points,m):
    M=len(points)
    N=len(points[0])
    Y_pred=[]
    for i in range(M):
        sum=0
        for j in range(N-1):
            sum+=m[j]*points[i,j]
        Y_pred.append(sum)
    return Y_pred
```

```python
def score(Y_o,Y_p):
    Y_o=np.array(Y_o)
    Y_p=np.array(Y_p)
    num=((Y_o-Y_p)**2).sum()
    den=((Y_o-Y_o.mean())**2).sum()
    ans=1-(num/den)
    return ans
```

```python
def cost(points,m):
    M=len(points)
    N=len(points[0])
    total_cost=0
    for i in range(M):
        y=points[i,N-1]
        sum=0
        for j in range(N-1):
            sum+=m[j]*points[i,j]
        ans=y-sum
        ans**=2
        total_cost+=ans
    return total_cost/M
        
```

```python
def gd(points,learning_rate,num_iterations):
    m=np.zeros(points.shape[1]-1,dtype=int)
    for i in range(num_iterations):
        m=step_gradient(points,learning_rate,m)
        #print(i,"Cost ",cost(points,m))
    return m
```

```python
def cost(X,Y,m):
    M=X.shape[0]
    N=X.shape[1]
    total_cost=0
    for i in range(M):
        y=Y[i]
        sum=0
        for j in range(N):
            sum+=m[j]*X[i,j]
        ans=y-sum
        ans**=2
        total_cost+=ans
    return total_cost/M
        
```

```python
def predict(X,m):
    M=X.shape[0]
    N=X.shape[1]
    Y_pred=[]
    for i in range(M):
        sum=0
        for j in range(N):
            sum+=m[j]*X[i,j]
        Y_pred.append(sum)
    return Y_pred
```

```python
def score(Y_o,Y_p):
    Y_o=np.array(Y_o)
    Y_p=np.array(Y_p)
    num=((Y_o-Y_p)**2).sum()
    den=((Y_o-Y_o.mean())**2).sum()
    ans=1-(num/den)
    return ans
```

```python
def step_gradient(X,Y,learning_rate,m):
    new_m=[]
    M=X.shape[0]
    N=X.shape[1]
    m_slope=[0 for i in range(N)]
    for i in range(M):
        y=Y[i]
        sum=0
        l=0
        for j in range(N):
            sum+=m[j]*X[i,j]
        for k in range(N):
            c=(-2/M)*(y-sum)*X[i,k]+m_slope[l]
            #print(c,end=" ")
            m_slope[l]=c 
            #print(m_slope[l]==c)
            l+=1
        
            
    for i in range(N):
        c=m[i]-learning_rate*m_slope[i]
        new_m.append(m[i]-learning_rate*m_slope[i])
    
    return new_m
```

```python
def gd(X,Y,learning_rate,num_iterations):
    m=np.zeros(X.shape[1],dtype=int)
    for i in range(num_iterations):
        m=step_gradient(X,Y,learning_rate,m)
        #print("i ",i,m)
        print(i,"Cost ",cost(X,Y,m))
    return m
```
