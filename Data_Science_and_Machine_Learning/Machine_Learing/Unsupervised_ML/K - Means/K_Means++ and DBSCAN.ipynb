```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles,make_moons
```

```python
X,Y=make_moons(n_samples=200,noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
#Here, the data is not linearly separable, it is non convex
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
centers=kmeans.cluster_centers_
pred=kmeans.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(centers[:,0],centers[:,1],c="red",marker="*")
plt.show()
# Here, kmeans are not able to predict the data
```

## DBSCAN

Density Based Spatial Clustering of Application with Noise

- Based upon the idea that the cluster is a high density area surrounded by low density area
- Starts by exploring a small area, if the density is good enough, it is considered to be the part of the cluster and explore neigbours and so on. 

It groups region based on density

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
#Here, the data is not linearly separable, it is non convex
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbs=DBSCAN(eps=0.2,min_samples=5)
dbs.fit(X)
```

```python
Ypred=dbs.fit_predict(X)
Ypred
```

```python
plt.scatter(X[:,0],X[:,1],c=Ypred)
plt.show()
#Here, the data is not linearly separable, it is non convex
```
