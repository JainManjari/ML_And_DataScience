## K-Means++

K-Means is quite sentive to the initialisation. If the init. is not good, then we don't get the desired no of clusters. To overcome this, we use K-Means++ technique, which chooses center as closely to the final ones.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
perf=cv2.imread("perf_clus.JPG")
perf=cv2.cvtColor(perf,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,8))
plt.title("Perfect Clustering by Using K-Means")
plt.imshow(perf)
plt.axis("off")
plt.show()
```

```python
nperf=cv2.imread("non_perf_clus.JPG")
nperf=cv2.cvtColor(nperf,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,8))
plt.title("Non Perfect Clustering by Using K-Means")
plt.imshow(nperf)
plt.axis("off")
plt.show()
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=500,centers=5,random_state=3,n_features=2)
```

```python
from sklearn.cluster import KMeans
```

```python
k_means=KMeans(n_clusters=5)
```

```python
k_means.fit(X,y)
```

```python
k_means.cluster_centers_
```

```python
centers=k_means.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(centers[:,0],centers[:,1],c="yellow",marker="*")
plt.show()
```

```python
pred=k_means.labels_
pred
```

```python
plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(centers[:,0],centers[:,1],c="red",marker="*")
plt.show()
```
