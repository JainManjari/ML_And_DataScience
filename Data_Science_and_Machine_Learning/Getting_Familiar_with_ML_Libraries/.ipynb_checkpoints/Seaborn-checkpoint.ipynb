```python
import seaborn as sns
import numpy as np
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#mean of the total_bills acquired by the male and the female
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.countplot(x="time",data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
#min,[mean] => The rang of max total_bill fall into, max, outlies
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue="smoker")
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue="sex")
```

```python
sns.boxplot(x='sex',y='total_bill',data=tips,hue="day")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
#same explaination as boxplot
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker",split=True)
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.distplot(tips["total_bill"],kde=False)
```

```python
#distribution plot
sns.distplot(tips["total_bill"],kde=False,bins=40)
```

```python
sns.distplot(tips["tip"],kde=False)
```

```python
#normal distribution
sns.kdeplot(tips["total_bill"])
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
#first dist plot of total_bill on top
#seconf dist plot of top on right
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind='hex')
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind='kde')
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
sns.pairplot(tips)
#pair wise relationships with every numeric feature
#1st row= total_bill+total_bill (normal distribution)| total_bill+tip |total_bill+size
#2nd row= tip+total_bill | tip+tip (normal distribution) | tip+size
#3rd row= size+total_bill | size+tip | size+size (normal distribution)
```

```python
sns.pairplot(tips,hue="sex")
```

```python
tips.head()
#Here we have varibales only on the columns=> total_bill,tip,sex etc.,.
#No variables exists on the rows => only indexes => 0,1,2, etc.,.
```

```python
tips_corr=tips.corr()
tips_corr
#correlation btw columns having numeric values
#correlations lies btw [-1,1] => -1 implies negative correlation => means if one item increase then other decreases
#1 implies positive correlation => means if one item increase then other also increases
```

```python
#tips_corr has variables on both rows and columns
sns.heatmap(tips_corr)
#lighter color higher correlation
```

```python
sns.heatmap(tips_corr,annot=True)
```
