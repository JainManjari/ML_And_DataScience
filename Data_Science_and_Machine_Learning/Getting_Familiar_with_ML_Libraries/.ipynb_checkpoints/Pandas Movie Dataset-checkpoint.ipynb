```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("movie_data.csv")
df.head()
```

```python
df.columns
```

```python
movies=df["movie_title"]
movies=list(np.array(movies))
movies
```

```python
d={}
for ele in movies:
    #l=ele.split(" ")
    n=len(ele)
    d[n]=d.get(n,0)+1
    
d
```

```python
x=list(d.keys())
x
```

```python
y=list(d.values())
y
```

```python
plt.style.use("seaborn")
plt.scatter(x,y)
plt.show()
```
