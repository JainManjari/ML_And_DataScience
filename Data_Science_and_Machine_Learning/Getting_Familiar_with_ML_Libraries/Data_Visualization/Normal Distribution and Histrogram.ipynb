# Generate a Normal Distribution for 1 D Array

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
#Standard Normal Distribution
u=0
sigma=1


vals=np.random.randn(1000) #retreieves the standard normalised values

std_norm=vals*sigma+u
```

```python
plt.hist(std_norm,bins=10)
plt.show()
```

```python
new_nd=vals*5+1
plt.hist(new_nd,bins=100)
plt.show()
```

```python
new_nd=np.round(new_nd)
z=np.unique(new_nd,return_counts=True)
z
#Here no lying closes to the mean or u (0), u-sigma=0-1=-1, u+sigma=0+1=1 has highest frequency
```

```python
x=np.random.randn(200)
x=x*3+5
y=np.zeros(x.shape)

plt.scatter(x,y)
plt.show()
```

# Mutlivariant Normal Distribution

```python
#Bivariate

# mean= [[mean(x),mean(y)]]


# cov=[  [sigma(x,x), sigma(x,y)],
#        [sigma(y,x), sigma(y,y)]
#     ]

#if sigma(x,x) is greater than sigma(y,y) => more spread on x axis or vice vera
#if sigma(x,y) = 0 = sigma(y,x) => then x and y are independent of each other

np.random.seed(0)

mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]]) 

dist=np.random.multivariate_normal(mean,cov,500)
```

```python
dist.shape
```

```python
plt.scatter(dist[:,0],dist[:,1])
plt.show()
#here the x and y are independent of each other and has even spread across x and y axis
```

```python
cov3=np.array([[1,0.8],[0.8,1]]) 
dist3=np.random.multivariate_normal(mean,cov3,500)

plt.scatter(dist3[:,0],dist3[:,1])
plt.show()
```

```python
mean2=np.array([5.0,6.0])
cov2=np.array([[3.0,0.2],[-10.8,10.1]]) 

dist2=np.random.multivariate_normal(mean2,cov2,500)

plt.scatter(dist3[:,0],dist3[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
