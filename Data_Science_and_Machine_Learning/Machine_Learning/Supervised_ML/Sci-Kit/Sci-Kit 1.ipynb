# Implementing Linear Regressing using SK

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
alg=LR()
```

```python
alg.fit(X,Y) #trainning the data
```

```python
alg.coef_ #theta1 and theta2
```

```python
alg.intercept_ #theta0
```

```python
alg.predict([X[0],X[1]]) #y0 and y1 predictions
```

```python
Y[0],Y[1] #actual values
```

```python
Y_pred=alg.predict(X) #predciting values
```
