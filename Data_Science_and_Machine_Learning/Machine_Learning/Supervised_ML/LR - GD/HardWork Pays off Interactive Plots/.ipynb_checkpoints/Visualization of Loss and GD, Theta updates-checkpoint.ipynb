```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import mpl_toolkits.mplot3d.axes3d
```

```python
X=pd.read_csv("Linear_X_Train.csv")
Y=pd.read_csv("Linear_Y_Train.csv")

X=X.values
Y=Y.values

#Normalisation

X=(X-X.mean())/X.std()
```

```python
# GD

def predict(X,m,c):
    return X*m+c

def score(Yo,Yp):
    n=np.sum((Yo-Yp)**2)
    d=np.sum((Yo-Yo.mean())**2)
    return (1-(n/d))*100

def cost(Yo,Yp):
    s=np.sum((Yo-Yp)**2)
    return s/Yo.shape[0]

def step_grad(X,Y,m,c,lr):
    m_slope=0
    c_slope=0
    M=X.shape[0]
    
    for i in range(M):
        x=X[i,0]
        y=Y[i,0]
        m_slope+=(-2/M)*(y-m*x-c)*(x)
        c_slope+=(-2/M)*(y-m*x-c)
    
    new_m=m-lr*m_slope
    new_c=c-lr*c_slope
    
    return new_m,new_c
    

def gradient_descent(X,Y,lr=0.1,iterations=100):
    m=0
    c=0
    error_list=[]
    theta_list=[]
    for i in range(iterations):
        m,c=step_grad(X,Y,m,c,lr)
        Y_pred=predict(X,m,c)
        error_list.append(cost(Y,Y_pred))
        theta_list.append((c,m))
        
    return m,c,error_list,theta_list
```

```python
m,c,error_list,theta_list=gradient_descent(X,Y)
```

# Loss Function

```python
# Loss Function


T0=np.arange(-40,40,1) #close to c
T1=np.arange(40,120) #close to m

T0,T1=np.meshgrid(T0,T1)

J=np.zeros(T0.shape)

for i in range(J.shape[0]):
    for j in range(J.shape[1]):
        yp=T1[i,j]*X+T0[i,j]
        J[i,j]=np.sum((Y-yp)**2)/Y.shape[0]
```

```python
J.shape
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(T0,T1,J,cmap="rainbow")
plt.show()
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.contour(T0,T1,J,cmap="rainbow")
plt.show()
```
