```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(10)
Y=(X-5)**2

print(X,Y)
```

# For given f(x), we will find x for which f(x) is minimised

```python
###Visualtization

plt.style.use("seaborn")
plt.plot(X,Y)
plt.ylabel("F(x)")
plt.xlabel("x")
plt.show()
```

```python
#intialised x 
x=0
lr=0.1 #learning rate
error=[]

#Moving first 50 steps downhill

plt.plot(X,Y)

for i in range(50):
    grad=2*(x-5) #diff of f(x)
    x=x-lr*grad
    print(x,end=" ") #here the answer is close to 5 like in graph
    y=(x-5)**2
    error.append(y)
    plt.scatter(x,y)
    
```

```python
#Plot the error

plt.plot(error)
plt.show()
```
