```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=datasets.load_boston()
print(type(data))
```

```python
X=data.data
Y=data.target
```

```python
df=pd.DataFrame(X)
df.columns=data.feature_names
print(df.columns)
```

```python
columns=df.columns
#Here, we are adding all the 2 degree columns in the dataset like CRIM_CRIM, CRIM_ZN, ZN_INDUS etc.,.
for i in range(len(columns)):
    for j in range(i,len(columns)):
        st=columns[i]+"_"+columns[j]
        df[st]=df[columns[i]]*df[columns[j]]

data2=df.values
X2=data2[:,:]
df
```

```python
df.describe()
```

```python
from sklearn import model_selection
```

```python
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,random_state=0)
X2_train,X2_test,Y2_train,Y2_test=model_selection.train_test_split(X2,Y,random_state=0)
#random state is used to send the same train data in X_train and X2_train and vice versa
```

```python
from sklearn.linear_model import LinearRegression

alg1=LinearRegression()
alg2=LinearRegression()
```

```python
alg1.fit(X_train,Y_train)
alg2.fit(X2_train,Y2_train)
```

```python
Y_predict=alg1.predict(X_test)
Y_predict2=alg2.predict(X2_test)
```

```python
print(alg1.score(X_train,Y_train)) #now adding the new colmn of age_age improves the score
print(alg2.score(X2_train,Y2_train)) #here, training data gets imporve
```

```python
print(alg1.score(X_test,Y_test))
print(alg2.score(X2_test,Y2_test)) #but the testing data gets reduced
```
