## Implementing Decision Tree

```python
# Define Entropy and Information Gain

def entropy(col):
    
    counts = np.unique(col,return_counts=True)
    N = float(col.shape[0])
    ent = 0.0
    for ix in counts[1]:
        p  = ix/N
        ent -= (p*np.log2(p))
    
    return ent
    


def divide_data(X,fkey,fval):
    #Work with Pandas Data Frames
    
    x_right = pd.DataFrame([],columns=X.columns)
    x_left = pd.DataFrame([],columns=X.columns)
    
    for ix in range(X.shape[0]):
        val = X[fkey].loc[ix]
        
        if val > fval:
            x_right = x_right.append(X.loc[ix])
        else:
            x_left = x_left.append(X.loc[ix])
            
    return x_left,x_right
    
#     left=X[X[fkey]<=fval]
#     right=X[X[fkey]>fval]
#     return left,right


def information_gain(x_data,fkey,fval):
    
    left,right = divide_data(x_data,fkey,fval)
    
    #% of total samples are on left and right
    l = float(left.shape[0])/x_data.shape[0]
    r = float(right.shape[0])/x_data.shape[0]
    
    #All examples come to one side!
    if left.shape[0] == 0 or right.shape[0] ==0:
        return -1000000 #Min Information Gain
    
    i_gain = entropy(x_data.Survived) - (l*entropy(left.Survived)+r*entropy(right.Survived))
    return i_gain
```

```python
class DecisionTree:
    
    #Constructor
    def __init__(self,depth=0,max_depth=5):
        self.left = None
        self.right = None
        self.fkey = None
        self.fval = None
        self.max_depth = max_depth
        self.depth = depth
        self.target = None
        
    def train(self,X_train):
        
        features = ['Pclass','Sex','Age','SibSp', 'Parch', 'Fare']
        info_gains = []
        
        for ix in features:
            i_gain = information_gain(X_train,ix,X_train[ix].mean())
            info_gains.append(i_gain)
            
        self.fkey = features[np.argmax(info_gains)]
        self.fval = X_train[self.fkey].mean()
        print("Making Tree Features is",self.fkey)
        
        #Split Data
        data_left,data_right = divide_data(X_train,self.fkey,self.fval)
        data_left = data_left.reset_index(drop=True)
        data_right = data_right.reset_index(drop=True)
         
        #Truly a left node
        if data_left.shape[0]  == 0 or data_right.shape[0] ==0:
            if X_train.Survived.mean() >= 0.5:
                self.target = "Survive"
            else:
                self.target = "Dead"
            return
        #Stop earyly when depth >=max depth
        if(self.depth>=self.max_depth):
            if X_train.Survived.mean() >= 0.5:
                self.target = "Survive"
            else:
                self.target = "Dead"
            return
        
        #Recursive Case
        self.left = DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.left.train(data_left)
        
        self.right = DecisionTree(depth=self.depth+1,max_depth=self.max_depth)
        self.right.train(data_right)
        
        #You can set the target at every node
        if X_train.Survived.mean() >= 0.5:
            self.target = "Survive"
        else:
            self.target = "Dead"
        return
    
    def predict(self,test):
        #Here test data also contains both X and Y test values together
        if test[self.fkey]>self.fval:
            #go to right
            if self.right is None:
                return self.target
            return self.right.predict(test)
        else:
            if self.left is None:
                return self.target
            return self.left.predict(test)
        
```

## Random Forests

- It is used to deal with high variance (the any tiny change in the data will lead to complete change of tree) of DT.
- It is to ensure overfitting doesn't occur

We will make many trees (like a forest). Then, we take average or majority of all the trees. Each tree will have different structure or features

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier(criterion="entropy")

dt.fit(X_train,Y_train)
dt.score(X_train,Y_train) #Here without mention of the max_depth, the accuracy have raised to 98% => overfitting on train data
```

```python
dt.score(X_test,Y_test) #it is reduced from 98% to 77%
```

```python
# One way to prevent overfitting

dt2=DecisionTreeClassifier(criterion="entropy",max_depth=5)

dt2.fit(X_train,Y_train)
dt2.score(X_train,Y_train), dt2.score(X_test,Y_test) # Here, the diff is very neligible. Hence, it is worth it
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier(n_estimators=10,criterion="entropy",max_depth=5)
rfc.fit(X_train,Y_train)
```

```python
rfc.score(X_train,Y_train) #overfitting can be eliminated by max_depth
```

```python
rfc.score(X_test,Y_test) 
```

```python
from sklearn.model_selection import cross_val_score


cv=cross_val_score(rfc,X_train,Y_train,cv=5).mean()
cv
```

```python
acc_list=[]

for i in range(1,100):
    r=RandomForestClassifier(n_estimators=i,criterion="entropy",max_depth=5)
    acc_list.append(cross_val_score(r,X_train,Y_train,cv=5).mean())
```

```python
acc_list=np.array(acc_list)
no=np.argmax(acc_list)+1 #no of the trees for highest accuracy
no
```

```python
import matplotlib.pyplot as plt

plt.style.use("seaborn")
plt.plot(acc_list)
plt.show()
```

```python
rfc=RandomForestClassifier(n_estimators=no,criterion="entropy",max_depth=5)
rfc.fit(X_train,Y_train)

rfc.score(X_train,Y_train),rfc.score(X_test,Y_test)
```
