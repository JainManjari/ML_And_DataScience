```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

img1=cv2.imread('D:/Python/DataScieneML/Supervised ML/Pokemon/Collage/Images/top_left.jpg')
rgb_img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img1=cv2.resize(rgb_img1,(200,200))

print(img1.shape)
plt.imshow(img1) #BGR Crop Image
plt.axis("off") #to remove axis
plt.show()
```

```python
img2=cv2.imread('D:/Python/DataScieneML/Supervised ML/Pokemon/Collage/Images/top_right.jpg')
rgb_img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img2=cv2.resize(rgb_img2,(200,200))

print(img2.shape)
plt.imshow(img2) #BGR Crop Image
plt.axis("off") #to remove axis
plt.show()
```

```python
i1=np.array(img1)
i2=np.array(img2)
#new_img=np.append(i1,i2,axis=1)
f1=np.hstack((i1,i2))


plt.imshow(f1) #BGR Crop Image
plt.axis("off") #to remove axis
plt.show()
```

```python
img3=cv2.imread('D:/Python/DataScieneML/Supervised ML/Pokemon/Collage/Images/bottom_left.jpg')
rgb_img3=cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
img3=cv2.resize(rgb_img3,(200,200))


print(img3.shape)
plt.imshow(img3) #BGR Crop Image
plt.axis("off") #to remove axis
plt.show()
```

```python
img4=cv2.imread('D:/Python/DataScieneML/Supervised ML/Pokemon/Collage/Images/bottom_right.jpg')
rgb_img4=cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
img4=cv2.resize(rgb_img4,(200,200))

print(img4.shape)
plt.imshow(img4) #BGR Crop Image
plt.axis("off") #to remove axis
plt.show()
```

```python
img5=cv2.imread('D:/Python/DataScieneML/Supervised ML/Pokemon/Collage/Images/center.jpeg')
rgb_img5=cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)
img5=cv2.resize(rgb_img5,(100,100))

bordersize = 10
img5 = cv2.copyMakeBorder(
    img5,
    top=bordersize,
    bottom=bordersize,
    left=bordersize,
    right=bordersize,
    borderType=cv2.BORDER_CONSTANT,
)

print(img5.shape)
plt.imshow(img5) #BGR Crop Image
#plt.title("Lion "+str(border.shape))|
plt.axis("off") #to remove axis
plt.show()
```

```python
b1=np.ones((200,10,3),dtype=int) #col,row,z
top=np.hstack((b1,img1,b1,img2,b1))
bottom=np.hstack((b1,img3,b1,img4,b1))

# plt.imshow(top)
# plt.show()

# plt.imshow(bottom)
# plt.show()

b2=np.zeros((10,430,3),dtype=int)
final=np.vstack((b2,top,b2,bottom,b2))
final[155:275,155:275]=img5 #col,row
plt.imshow(final)
plt.show()
```
