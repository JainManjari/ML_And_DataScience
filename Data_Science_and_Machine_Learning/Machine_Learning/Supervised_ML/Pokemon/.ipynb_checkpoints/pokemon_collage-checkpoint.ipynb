```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os
```

```python
path="D:/Python/ML/Pokemon/Collage/Images"
imgs_dir=os.listdir(path)
imgs_list=[]

for img in imgs_dir:
    print(img)
    p=path+"/"+img
    im=cv2.imread(p)
    im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
    im=cv2.resize(im,(200,200))
    imgs_list.append(im)
```

```python
for im in imgs_list:
    plt.imshow(im)
    plt.show()
```

```python
center=imgs_list[2]
center=cv2.resize(center,(100,100))
center=cv2.copyMakeBorder(center, 10, 10, 10, 10, cv2.BORDER_CONSTANT) 
imgs_list[2]=center
```

```python
border=np.zeros((200,10,3),dtype=int)
bottom=np.hstack((border,imgs_list[0],border,imgs_list[1],border))
top=np.hstack((border,imgs_list[3],border,imgs_list[4],border))


border2=np.ones((10,430,3),dtype=int)

collage=np.vstack((border2,top,border2,bottom,border2))
collage[155:275,155:275]=imgs_list[2]

plt.imshow(collage)
plt.show()
```

```python
data=collage.copy()
data=data.reshape(-1,3)
df=pd.DataFrame(data)
df.columns=['r','g','b']
df
```

```python
df.to_csv('D:/Python/ML/Pokemon/Collage/pokemon_collage.csv',index=False)
```
